```python
import pandas as pd
import numpy as np
import os
import re
from pathlib import Path
from fileinput import filename
from matplotlib import lines
```

```python
PROJECT_ROOT = Path().resolve().parents[0]
DATA_DIR = PROJECT_ROOT / "data"
_1999_df = pd.read_csv(DATA_DIR / "raw/cohort_1999_2005.csv")
_2005_df = pd.read_csv(DATA_DIR / "raw/cohort_2005_2019.csv")

# Rename columns in 2005+ dataset to match the older dataset
_2005_df = _2005_df.rename(columns={
    "player_id": "PERSON_ID",
    "player_name": "PLAYER_NAME",
    "draft_year": "SEASON"
})

# Combine and remove exact duplicate rows
combined_df = pd.concat([_1999_df, _2005_df], ignore_index=True).drop_duplicates()

# Save
output_path = DATA_DIR / "cohort_1999_2019.csv"
combined_df.to_csv(output_path, index=False)

print(f"Saved combined dataset to: {output_path}")
print(f"Final shape: {combined_df.shape}")
```
